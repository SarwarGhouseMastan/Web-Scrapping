```python
import pandas as pd
import numpy as np
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
webpage =requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage , 'lxml')
soup
```

```python
#Cleaning the above output
soup.prettify()
```

```python
#Extracting the title
soup.find_all('h1')
```

```python
#removing the unwanted part of the title and extracting the title only (code=h1)
soup.find_all('h1')[0].text
```

```python
#Extracting the companies , by looking at the page source from the website (h2---> is the key for the company names) 
# (code=h2)
soup.find_all('h2')
```

```python
#Removing the unwanted part , and finding the companies names only
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
#Extract from the page source(code=p)
for i in soup.find_all('p'):
    print(i.text.strip())
```

```python
#Extract Ratings from the page source(code=a)
soup.find_all('a')
```

```python
# No of companies in one page
len(soup.find_all('a',class_ = 'review-count' ))
```

```python
#Considering whole content container
company = soup.find_all('div' , class_='company-content-wrapper')
company
```

```python
len(company)
```

```python
name = []
rating = []
reviews = []
ctype = []
hq = []
how_old = []
no_of_employees = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p' , class_='rating').text.strip())
    reviews.append(i.find('a' , class_='review-count').text.strip())
    ctype.append(i.find_all('p', class_='infoEntity' )[0].text.strip())
    hq.append(i.find_all('p' , class_='infoEntity')[1].text.strip())
    how_old.append(i.find_all('p', class_='infoEntity' )[2].text.strip())
    no_of_employees.append(i.find_all('p', class_='infoEntity' )[3].text.strip())
```

```python
#Creating the Data frame from the created lists
df = pd.DataFrame({'name': name , 'rating': rating, 'reviews': reviews , 'company type': ctype , 'headquarters': hq, 
                   'company age': how_old , 'No of Employees working': no_of_employees})
df
```

```python
#Saving to CSV format
df.to_csv('Ambitionbox.csv')
```
